```python
# Estimate intrinsic timescales τ via autocorrelation
def estimate_tau(x):
    ac = np.correlate(x - np.mean(x), x - np.mean(x), mode='full')
    ac = ac[ac.size // 2:]
    ac /= ac[0]
    try:
        idx = np.where(ac < 1/np.e)[0][0]
        return idx
    except IndexError:
        return np.nan
    
def run_esn(inputs):
    # set up
    
    # -----------------------------
    # Build the Echo State Network (ESN)
    n_inputs = inputs.shape[0]
    n_reservoir = 1000
    n_input_driven = 300
    spectral_radius = 1
    leak_rate = 0.2
    lambda_decay = 300  # Decay scale for recurrent connections

    #np.random.seed(42)

    # Input weights: only project to first 300 neurons
    Win = np.zeros((n_reservoir, n_inputs))
    Win[:n_input_driven, :] = np.random.uniform(-1, 1, (n_input_driven, n_inputs)) * 0.5

    # Recurrent weights with exponential distance decay
    indices = np.arange(n_reservoir)
    D = np.abs(indices[:, None] - indices[None, :])  # Distance matrix
    E = np.exp(-D / lambda_decay)                   # Exponential decay matrix
    W_raw = np.random.randn(n_reservoir, n_reservoir)
    W = W_raw * E
    # Normalize spectral radius
    eigvals = np.linalg.eigvals(W)
    W *= spectral_radius / np.max(np.abs(eigvals)) #* 1.5

    
    # run
    
    
    x = np.zeros(n_reservoir)
    states = []
    for t in range(inputs.shape[1]):
        u = inputs[:, t]
        x = (1 - leak_rate) * x + leak_rate * np.tanh(W @ x + Win @ u)
        states.append(x.copy())
    return np.array(states).T  # shape: (n_reservoir, time_steps)

```

```python

# -------------------------
# Plotting
fig, axs = plt.subplots(2, 3, figsize=(16, 9))
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr, wilcoxon

# -------------------------
# Config
n_input_dim = 50
time_steps = 2682
n_reservoir = 1000
n_input_driven = 300
n_runs = 2
seed = 3

# -------------------------
# Input generators

def generate_rest_shuffled(inputs):
    shuffled = inputs.copy()
    time_idx = np.arange(time_steps)
    np.random.shuffle(time_idx)
    return shuffled[:, time_idx]

import numpy as np

def bin_shuffle(inputs, bin_size):
    """
    Shuffles the input array in temporal bins of size `bin_size`.

    Parameters:
        inputs (np.ndarray): Array of shape (features, time_steps), e.g., (50, 2682)
        bin_size (int): Size of each bin to shuffle

    Returns:
        np.ndarray: The bin-shuffled array of the same shape
    """
    features, time_steps = inputs.shape
    num_bins = time_steps // bin_size

    # Trim to full bins only (optional: pad instead if needed)
    trimmed_length = num_bins * bin_size
    trimmed_inputs = inputs[:, :trimmed_length]

    # Reshape into bins: (features, num_bins, bin_size)
    binned = trimmed_inputs.reshape(features, num_bins, bin_size)

    # Shuffle the bins along the second axis (num_bins)
    shuffled_bin_indices = np.random.permutation(num_bins)
    binned_shuffled = binned[:, shuffled_bin_indices, :]

    # Reshape back to (features, trimmed_length)
    shuffled = binned_shuffled.reshape(features, trimmed_length)

    # If original length not divisible by bin_size, append the remaining time points
    if trimmed_length < time_steps:
        remaining = inputs[:, trimmed_length:]
        shuffled = np.concatenate([shuffled, remaining], axis=1)

    return shuffled


# -------------------------
# Analysis function
def run_condition(inputs):
    activity = run_esn(inputs)
    taus = np.array([estimate_tau(activity[i]) for i in range(n_reservoir)])
    fc = np.corrcoef(activity)
    mean_fc = np.nanmean(np.abs(fc), axis=1)
    return taus, mean_fc

# -------------------------
# Initialize containers
taus_all = {"Task": [], "Rest-Random": [], "Rest-Shuffled": []}
fc_all   = {"Task": [], "Rest-Random": [], "Rest-Shuffled": []}
r_vals   = {"Task": [], "Rest-Random": [], "Rest-Shuffled": []}

# New containers for full data
taus_all_full = {"Task": [], "Rest-Random": [], "Rest-Shuffled": []}
fc_all_full   = {"Task": [], "Rest-Random": [], "Rest-Shuffled": []}

# -------------------------
# Fixed Task input
np.random.seed(42)
inputs_task = inputs  # <-- define externally
#rest_random = randomized  # assume these are defined externally
#rest_shuff = generate_rest_shuffled(inputs_task)

rest_shuff = bin_shuffle(inputs_task,100)
rest_random = bin_shuffle(inputs_task,1)

# -------------------------
# Run loop
for i in range(n_runs):
    print(f"\n--- Run {i+1} ---")

    np.random.seed(i)
    taus_task, fc_task = run_condition(inputs_task)
    np.random.seed(i)
    taus_rand, fc_rand = run_condition(rest_random)
    np.random.seed(i)
    taus_shuff, fc_shuff = run_condition(rest_shuff)

    for cond, taus, fc in zip(["Task", "Rest-Random", "Rest-Shuffled"],
                              [taus_task, taus_rand, taus_shuff],
                              [fc_task, fc_rand, fc_shuff]):

        # Store full τ and FC values
        taus_all_full[cond].append(taus)
        fc_all_full[cond].append(fc)

        # Filter out input-driven units (first 300)
        valid = (~np.isnan(taus)) & (np.arange(n_reservoir) >= n_input_driven)
        taus_all[cond].append(taus[valid])
        fc_all[cond].append(fc[valid])

        if np.sum(valid) >= 2:
            r, _ = pearsonr(taus[valid], fc[valid])
        else:
            r = np.nan
        r_vals[cond].append(r)

    print(f"r (Task) = {r_vals['Task'][-1]:.2f}, "
          f"r (Rest-Random) = {r_vals['Rest-Random'][-1]:.2f}, "
          f"r (Rest-Shuffled) = {r_vals['Rest-Shuffled'][-1]:.2f}")

# -------------------------
# Convert to arrays
r_arrs = {k: np.array(v) for k, v in r_vals.items()}

# -------------------------
# Per-run Wilcoxon tests on τ and FC distributions
tau_wilcoxon_stats, fc_wilcoxon_stats = {"Rand": [], "Shuff": []}, {"Rand": [], "Shuff": []}
tau_pvals, fc_pvals = {"Rand": [], "Shuff": []}, {"Rand": [], "Shuff": []}

for i in range(n_runs):
    tau_task = taus_all["Task"][i]
    fc_task = fc_all["Task"][i]
    tau_rand = taus_all["Rest-Random"][i]
    fc_rand = fc_all["Rest-Random"][i]
    tau_shuff = taus_all["Rest-Shuffled"][i]
    fc_shuff = fc_all["Rest-Shuffled"][i]

    # τ
    stat, p = wilcoxon(tau_task, tau_rand, zero_method="wilcox", correction=True)
    tau_wilcoxon_stats["Rand"].append(stat)
    tau_pvals["Rand"].append(p)

    stat, p = wilcoxon(tau_task, tau_shuff, zero_method="wilcox", correction=True)
    tau_wilcoxon_stats["Shuff"].append(stat)
    tau_pvals["Shuff"].append(p)

    # FC
    stat, p = wilcoxon(fc_task, fc_rand, zero_method="wilcox", correction=True)
    fc_wilcoxon_stats["Rand"].append(stat)
    fc_pvals["Rand"].append(p)

    stat, p = wilcoxon(fc_task, fc_shuff, zero_method="wilcox", correction=True)
    fc_wilcoxon_stats["Shuff"].append(stat)
    fc_pvals["Shuff"].append(p)

# -------------------------
# Print summary
def summary(arr): return np.mean(arr), np.std(arr)

print("\n--- τ–FC correlation (mean ± SD) ---")
for cond, arr in r_arrs.items():
    m, s = summary(arr)
    print(f"{cond:14s}: r = {m:.3f} ± {s:.3f}")

def summarize_pvals(pvals, label):
    pvals = np.array(pvals)
    sig_count = np.sum(pvals < 0.05)
    return f"{label}: mean p = {np.mean(pvals):.2e}, median = {np.median(pvals):.2e}, {sig_count}/{len(pvals)} significant (p<0.05)"

print("\n--- Per-run Wilcoxon tests on full distributions ---")
print(summarize_pvals(tau_pvals["Rand"],  "τ  (Task vs Rest-Random)"))
print(summarize_pvals(fc_pvals["Rand"],   "FC (Task vs Rest-Random)"))
print(summarize_pvals(tau_pvals["Shuff"], "τ  (Task vs Rest-Shuffled)"))
print(summarize_pvals(fc_pvals["Shuff"],  "FC (Task vs Rest-Shuffled)"))

# τ–FC correlation
axs[0, 0].boxplot([r_arrs["Task"], r_arrs["Rest-Random"], r_arrs["Rest-Shuffled"]],
                  labels=["Task", "Rest-Random", "Rest-Shuffled"])
axs[0, 0].set_ylabel("τ–FC Correlation (r)")
axs[0, 0].set_title("τ–FC Correlation")

# Mean τ
taus_means = {k: [np.mean(x) for x in v] for k, v in taus_all.items()}
axs[0, 1].boxplot([taus_means["Task"], taus_means["Rest-Random"], taus_means["Rest-Shuffled"]],
                  labels=["Task", "Rest-Random", "Rest-Shuffled"])
axs[0, 1].set_ylabel("Mean τ")
axs[0, 1].set_title("Mean Intrinsic Timescale")

# Mean FC
fc_means = {k: [np.mean(x) for x in v] for k, v in fc_all.items()}
axs[0, 2].boxplot([fc_means["Task"], fc_means["Rest-Random"], fc_means["Rest-Shuffled"]],
                  labels=["Task", "Rest-Random", "Rest-Shuffled"])
axs[0, 2].set_ylabel("Mean |FC|")
axs[0, 2].set_title("Mean Functional Connectivity")

# τ p-value violins
axs[1, 0].violinplot([tau_pvals["Rand"], tau_pvals["Shuff"]])
axs[1, 0].set_xticks([1, 2])
axs[1, 0].set_xticklabels(["Rand", "Shuff"])
axs[1, 0].set_ylabel("Wilcoxon p-value (τ)")
axs[1, 0].set_title("p-values (Wilcoxon on τ)")

# FC p-value violins
axs[1, 1].violinplot([fc_pvals["Rand"], fc_pvals["Shuff"]])
axs[1, 1].set_xticks([1, 2])
axs[1, 1].set_xticklabels(["Rand", "Shuff"])
axs[1, 1].set_ylabel("Wilcoxon p-value (FC)")
axs[1, 1].set_title("p-values (Wilcoxon on FC)")

# Histograms: example from one run
idx = 0  # show run 1
axs[1, 2].hist(taus_all["Task"][idx], bins=30, alpha=0.6, label="Task", density=True)
axs[1, 2].hist(taus_all["Rest-Random"][idx], bins=30, alpha=0.6, label="Rest-Random", density=True)
axs[1, 2].hist(taus_all["Rest-Shuffled"][idx], bins=30, alpha=0.6, label="Rest-Shuffled", density=True)
axs[1, 2].set_title("τ Distributions (Run 1)")
axs[1, 2].set_xlabel("τ")
axs[1, 2].legend()

# Final layout
for ax in axs.flat:
    ax.grid(True)

plt.suptitle("Effect of Input Type on Reservoir Timescales and FC", fontsize=16)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import ranksums
import itertools

n_reservoir = 1000
split_index = 300
conditions = ["Task", "Rest-Shuffled", "Rest-Random"]
cond_colors = ["#1f77b4", "#2ca02c", "#ff7f0e"]

# Store τ medians per region
region_medians = {
    "0-299": {cond: [] for cond in conditions},
    "300-999": {cond: [] for cond in conditions}
}

for cond in conditions:
    taus_runs = taus_all_full[cond]  # list of arrays, each length 1000

    for run_taus in taus_runs:
        taus_full = np.array(run_taus)

        # Region 0–299
        region1 = taus_full[:split_index]
        region1 = region1[~np.isnan(region1)]
        region_medians["0-299"][cond].append(np.median(region1))

        # Region 300–999
        region2 = taus_full[split_index:]
        region2 = region2[~np.isnan(region2)]
        region_medians["300-999"][cond].append(np.median(region2))

# Plot
fig, axs = plt.subplots(1, 2, figsize=(13, 6), sharey=False)

for ax, region in zip(axs, ["0-299", "300-999"]):
    data = [region_medians[region][cond] for cond in conditions]
    
    # Boxplot
    bp = ax.boxplot(data, patch_artist=True,
                    boxprops=dict(facecolor='lightblue'),
                    medianprops=dict(color='black', linewidth=2.5),
                    showfliers=False)

    for patch, color in zip(bp['boxes'], cond_colors):
        patch.set_facecolor(color)

    ax.set_xticks([1, 2, 3])
    ax.set_xticklabels(conditions, rotation=20)
    ax.set_title(f"Median τ per Run\nNeurons {region}")
    ax.set_ylabel("τ")
    ax.grid(True)

    # Add statistical comparisons
    pairs = list(itertools.combinations(range(3), 2))
    height = max([max(d) for d in data]) * 1.05
    step = (max([max(d) for d in data]) - min([min(d) for d in data])) * 0.1
    for i, (i1, i2) in enumerate(pairs):
        stat, p = ranksums(data[i1], data[i2])
        label = f"p = {p:.3f}" if p >= 0.001 else "p < 0.001"
        x1, x2 = i1 + 1, i2 + 1
        y = height + i * step
        ax.plot([x1, x1, x2, x2], [y, y + 0.01, y + 0.01, y], lw=1.2, c='k')
        ax.text((x1 + x2) * 0.5, y + 0.01, label, ha='center', va='bottom', fontsize=9)

plt.suptitle("Median τ across Runs by Reservoir Segment\n(with Wilcoxon p-values)", fontsize=16)
plt.tight_layout(rect=[0, 0.05, 1, 0.95])
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import kruskal, mannwhitneyu
import itertools

# Approximate Entropy
def approximate_entropy(U, m=2, r=None):
    if r is None:
        r = 0.2 * np.std(U)
    N = len(U)

    def _phi(m):
        x = np.array([U[i:i + m] for i in range(N - m + 1)])
        C = np.sum(np.max(np.abs(x[:, None] - x[None, :]), axis=2) <= r, axis=0) / (N - m + 1)
        return np.sum(np.log(C + 1e-10)) / (N - m + 1)

    return abs(_phi(m) - _phi(m + 1))

# Sliding-window entropy per 1D signal
def sliding_window_entropy(U, win_size=100, step=10, m=2):
    N = len(U)
    entropies = []
    for start in range(0, N - win_size + 1, step):
        window = U[start:start + win_size]
        apen = approximate_entropy(window, m=m)
        entropies.append(apen)
    return np.array(entropies)

# Apply to multi-channel input
def compute_sliding_entropy_all_channels(data, win_size=100, step=10, m=2):
    n_channels = data.shape[0]
    results = []
    for i in range(n_channels):
        ent = sliding_window_entropy(data[i], win_size=win_size, step=step, m=m)
        results.append(ent)
    return np.array(results)  # shape: (n_channels, n_windows)

# Compute entropy for each condition
print("start first")
entropy_task = compute_sliding_entropy_all_channels(inputs_task, win_size=100, step=10, m=2)
print("start_second")
entropy_shuff = compute_sliding_entropy_all_channels(rest_shuff, win_size=100, step=10, m=2)
print("start third")
entropy_rand = compute_sliding_entropy_all_channels(randomized, win_size=100, step=10, m=2)

# Average across channels
mean_task = np.mean(entropy_task, axis=0)
mean_shuff = np.mean(entropy_shuff, axis=0)
mean_rand = np.mean(entropy_rand, axis=0)

# Plot time course
x = np.arange(len(mean_task)) * 10 + 50  # center of each window
plt.figure(figsize=(10,5))
plt.plot(x, mean_task, label="Task")
plt.plot(x, mean_shuff, label="Rest-Shuffled")
plt.plot(x, mean_rand, label="Rest-Random")
plt.xlabel("Sample index")
plt.ylabel("Approximate Entropy (m=2)")
plt.title("Sliding Window Entropy Across Conditions")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Boxplot of all entropy values (flatten across channels and windows)
flat_task = entropy_task.flatten()
flat_shuff = entropy_shuff.flatten()
flat_rand = entropy_rand.flatten()

plt.figure(figsize=(8,5))
plt.boxplot([flat_task, flat_shuff, flat_rand], labels=["Task", "Rest-Shuffled", "Rest-Random"])
plt.ylabel("Approximate Entropy")
plt.title("Entropy Distribution Across Conditions")
plt.grid(True)
plt.tight_layout()
plt.show()

# Statistical comparison
print("=== Kruskal-Wallis Test ===")
H, p = kruskal(flat_task, flat_shuff, flat_rand)
print(f"H = {H:.3f}, p = {p:.4e}")

print("\n=== Pairwise Mann–Whitney U Tests ===")
labels = ["Task", "Rest-Shuffled", "Rest-Random"]
groups = [flat_task, flat_shuff, flat_rand]
for (i, j) in itertools.combinations(range(3), 2):
    u_stat, pval = mannwhitneyu(groups[i], groups[j], alternative='two-sided')
    print(f"{labels[i]} vs {labels[j]}: U = {u_stat:.1f}, p = {pval:.4e}")
```

```python
import numpy as np
from sklearn.model_selection import StratifiedKFold, GridSearchCV, cross_val_predict
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# -------------------------
# Reshape tau data into feature matrix
# Each row = one run for one condition (vector of 700 taus)
X, y = [], []
label_map = {"Task": 0, "Rest-Shuffled": 1, "Rest-Random": 2}

for cond in ["Task", "Rest-Shuffled", "Rest-Random"]:
    for taus in taus_all[cond]:
        if len(taus) == 700:  # sanity check
            X.append(taus)
            y.append(label_map[cond])
        else:
            print(f"Warning: Skipped one run in {cond} with {len(taus)} taus")

X = np.array(X)
y = np.array(y)

print(f"Feature matrix shape: {X.shape} (samples x features)")
print(f"Labels shape: {y.shape}")

# -------------------------
# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# -------------------------
# Cross-validation and SVM grid search
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
param_grid = {
    'C': [0.1, 1, 10],
    'gamma': [0.001, 0.01, 0.1],
    'kernel': ['rbf']
}

grid = GridSearchCV(SVC(), param_grid, cv=cv, scoring='accuracy')
grid.fit(X_scaled, y)

# Report best model
print("Best parameters from grid search:", grid.best_params_)
print("Best cross-validated accuracy: {:.2f}".format(grid.best_score_))

# -------------------------
# Predict and plot confusion matrix
best_clf = grid.best_estimator_
y_pred = cross_val_predict(best_clf, X_scaled, y, cv=cv)

cm = confusion_matrix(y, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm,
                              display_labels=["Task", "Shuffled", "Random"])
disp.plot(cmap='Blues')
plt.title("Confusion Matrix: SVM on τ Features")
plt.show()
```

```python
#be sure to reload the data before running this
import numpy as np
from sklearn.model_selection import StratifiedKFold
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler, label_binarize
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, roc_curve, auc
import matplotlib.pyplot as plt

# -------------------------
# --- Load / prepare tau data ---
# Assuming taus_all is already loaded in memory: dict with keys: "Task", "Rest-Shuffled", "Rest-Random"
X, y = [], []
label_map = {"Task": 0, "Rest-Shuffled": 1, "Rest-Random": 2}

for cond in ["Task", "Rest-Shuffled", "Rest-Random"]:
    for taus in taus_all[cond]:
        if len(taus) == 700:  # sanity check
            X.append(taus)
            y.append(label_map[cond])
        else:
            print(f"Warning: Skipped one run in {cond} with {len(taus)} taus")

X = np.array(X)
y = np.array(y)

print(f"Feature matrix shape: {X.shape} (samples x features)")
print(f"Labels shape: {y.shape}")

# -------------------------
# --- Nested CV function ---
def run_nested_cv_tau(X, y, param_grid, cv_outer=5, cv_inner=3):
    scaler = StandardScaler()
    X = scaler.fit_transform(X)

    outer_cv = StratifiedKFold(n_splits=cv_outer, shuffle=True, random_state=42)
    inner_cv = StratifiedKFold(n_splits=cv_inner, shuffle=True, random_state=1)

    y_true_all, y_pred_all, y_proba_all = [], [], []

    for train_idx, test_idx in outer_cv.split(X, y):
        X_train, y_train = X[train_idx], y[train_idx]
        X_test, y_test = X[test_idx], y[test_idx]

        best_score = -np.inf
        best_params = None

        # Inner loop grid search
        for C in param_grid['C']:
            for gamma in param_grid['gamma']:
                clf = SVC(C=C, gamma=gamma, kernel='rbf', probability=True)
                scores = []
                for inner_train, inner_val in inner_cv.split(X_train, y_train):
                    clf.fit(X_train[inner_train], y_train[inner_train])
                    scores.append(clf.score(X_train[inner_val], y_train[inner_val]))
                mean_score = np.mean(scores)
                if mean_score > best_score:
                    best_score = mean_score
                    best_params = {'C': C, 'gamma': gamma}

        # Train best model on full train set
        best_clf = SVC(**best_params, kernel='rbf', probability=True)
        best_clf.fit(X_train, y_train)

        y_pred = best_clf.predict(X_test)
        y_proba = best_clf.predict_proba(X_test)

        y_true_all.append(y_test)
        y_pred_all.append(y_pred)
        y_proba_all.append(y_proba)

    y_true_all = np.concatenate(y_true_all)
    y_pred_all = np.concatenate(y_pred_all)
    y_proba_all = np.vstack(y_proba_all)

    # Confusion matrix (normalized)
    cm = confusion_matrix(y_true_all, y_pred_all)
    cm_norm = cm.astype(float) / cm.sum(axis=1, keepdims=True)

    # Binarize labels for ROC
    y_true_bin = label_binarize(y_true_all, classes=np.unique(y))

    return cm_norm, y_true_bin, y_proba_all, y_true_all, y_pred_all

# -------------------------
# --- ROC computation ---
def compute_roc_ci(y_true_bin, y_proba, n_bootstraps=1000, ci=0.95):
    n_classes = y_true_bin.shape[1]
    fpr, tpr, roc_auc = {}, {}, {}
    tpr_lower, tpr_upper = {}, {}

    for i in range(n_classes):
        fpr[i], tpr[i], _ = roc_curve(y_true_bin[:, i], y_proba[:, i])
        roc_auc[i] = auc(fpr[i], tpr[i])

        rng = np.random.RandomState(42)
        boot_tprs = []
        base_fpr = np.linspace(0, 1, 100)
        interp_tpr = np.interp(base_fpr, fpr[i], tpr[i])

        for _ in range(n_bootstraps):
            indices = rng.randint(0, len(y_proba), len(y_proba))
            if len(np.unique(y_true_bin[indices, i])) < 2:
                continue
            fpr_bs, tpr_bs, _ = roc_curve(y_true_bin[indices, i], y_proba[indices, i])
            interp_tpr_bs = np.interp(base_fpr, fpr_bs, tpr_bs)
            boot_tprs.append(interp_tpr_bs)
        boot_tprs = np.array(boot_tprs)

        lower = np.percentile(boot_tprs, ((1 - ci) / 2) * 100, axis=0)
        upper = np.percentile(boot_tprs, (1 - (1 - ci) / 2) * 100, axis=0)

        tpr_lower[i] = lower
        tpr_upper[i] = upper
        fpr[i] = base_fpr
        tpr[i] = interp_tpr

    return fpr, tpr, roc_auc, tpr_lower, tpr_upper

# -------------------------
# --- Run classification ---
param_grid = {'C': [0.1, 1, 10], 'gamma': [0.001, 0.01, 0.1]}
cm_norm, y_true_bin, y_proba, y_true_all, y_pred_all = run_nested_cv_tau(X, y, param_grid)

# -------------------------
# --- Plot ---
fig, axs = plt.subplots(1, 2, figsize=(14, 6))

# Confusion matrix
disp = ConfusionMatrixDisplay(confusion_matrix=cm_norm, display_labels=["Task", "Shuffled", "Random"])
disp.plot(ax=axs[0], cmap="Blues", colorbar=False)
axs[0].set_title("Normalized Confusion Matrix")

# ROC curves
fpr, tpr, roc_auc, tpr_lower, tpr_upper = compute_roc_ci(y_true_bin, y_proba)
colors = ['blue', 'green', 'red']
labels = ["Task", "Shuffled", "Random"]

for i, color in enumerate(colors):
    axs[1].plot(fpr[i], tpr[i], color=color, label=f"{labels[i]} (AUC = {roc_auc[i]:.2f})")
    axs[1].fill_between(fpr[i], tpr_lower[i], tpr_upper[i], color=color, alpha=0.2)

axs[1].plot([0, 1], [0, 1], 'k--', alpha=0.5)
axs[1].set_xlabel("False Positive Rate")
axs[1].set_ylabel("True Positive Rate")
axs[1].set_title("Multi-class ROC Curves with 95% CI")
axs[1].legend(loc="lower right")
axs[1].grid(True)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics.pairwise import cosine_similarity
from scipy.stats import mannwhitneyu
from statsmodels.stats.multitest import multipletests

def mean_bin_embeddings(embeddings, bin_size=100):
    features, time_steps = embeddings.shape
    num_bins = time_steps // bin_size
    trimmed = embeddings[:, :num_bins * bin_size]
    reshaped = trimmed.reshape(features, num_bins, bin_size)
    return reshaped.mean(axis=2)  # shape (features, num_bins)

def adjacent_bin_similarity(bin_means):
    sims = []
    for i in range(bin_means.shape[1] - 1):
        sim = cosine_similarity(bin_means[:, i].reshape(1, -1), bin_means[:, i+1].reshape(1, -1))[0,0]
        sims.append(sim)
    return np.array(sims)

def smooth(x, win=5):
    return np.convolve(x, np.ones(win)/win, mode='same')

# Parameters
bin_size = 100
smooth_win = 5

# Compute bin means and adjacent similarities
task_bin_means = mean_bin_embeddings(inputs_task, bin_size)
shuff_bin_means = mean_bin_embeddings(rest_shuff, bin_size)
rand_bin_means = mean_bin_embeddings(rest_random, bin_size)

task_adj_sims = adjacent_bin_similarity(task_bin_means)
shuff_adj_sims = adjacent_bin_similarity(shuff_bin_means)
rand_adj_sims = adjacent_bin_similarity(rand_bin_means)

# Smooth for visualization
task_smooth = smooth(task_adj_sims, smooth_win)
shuff_smooth = smooth(shuff_adj_sims, smooth_win)
rand_smooth = smooth(rand_adj_sims, smooth_win)

# Statistical test: Mann-Whitney U between pairs
pairs = [
    ('Task', task_adj_sims, 'Rest shuffled', shuff_adj_sims),
    ('Task', task_adj_sims, 'Rest random', rand_adj_sims),
    ('Rest shuffled', shuff_adj_sims, 'Rest random', rand_adj_sims),
]

pvals = []
comparisons = []
for (g1, d1, g2, d2) in pairs:
    stat, p = mannwhitneyu(d1, d2, alternative='two-sided')
    pvals.append(p)
    comparisons.append(f"{g1} vs {g2}")

# FDR correction
reject, pvals_corrected, _, _ = multipletests(pvals, alpha=0.05, method='fdr_bh')

# Plotting
fig, axes = plt.subplots(1, 2, figsize=(14,5))

# Left: smooth adjacent-bin similarity time series
axes[0].plot(task_smooth, label='Task')
axes[0].plot(shuff_smooth, label='Rest shuffled')
axes[0].plot(rand_smooth, label='Rest random')
axes[0].set_xlabel('Bin index')
axes[0].set_ylabel('Cosine similarity between adjacent bins')
axes[0].set_title('Smoothed adjacent-bin similarity over narrative')
axes[0].legend()

# Right: boxplot of adjacent-bin similarity distributions
data = [task_adj_sims, shuff_adj_sims, rand_adj_sims]
labels = ['Task', 'Rest shuffled', 'Rest random']
axes[1].boxplot(data, labels=labels)
axes[1].set_ylabel('Cosine similarity between adjacent bins')
axes[1].set_title('Distribution of adjacent-bin similarities')

# Annotate significance on boxplot
y_max = max([max(d) for d in data])
y_min = min([min(d) for d in data])
y_step = (y_max - y_min) * 0.1
h = y_max + y_step
for i, (comp, pval, sig) in enumerate(zip(comparisons, pvals_corrected, reject)):
    if sig:
        g1, g2 = pairs[i][0], pairs[i][2]
        idx1, idx2 = labels.index(g1) + 1, labels.index(g2) + 1
        axes[1].plot([idx1, idx1, idx2, idx2], [h, h+0.02, h+0.02, h], lw=1.5, c='k')
        axes[1].text((idx1+idx2)/2, h+0.025, "*", ha='center', va='bottom', fontsize=14)
        h += y_step

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics.pairwise import cosine_similarity

def mean_bin_embeddings(embeddings, bin_size=100):
    features, time_steps = embeddings.shape
    num_bins = time_steps // bin_size
    trimmed = embeddings[:, :num_bins * bin_size]
    reshaped = trimmed.reshape(features, num_bins, bin_size)
    return reshaped.mean(axis=2)  # shape (features, num_bins)

def adjacent_bin_similarity(bin_means):
    sims = []
    for i in range(bin_means.shape[1] - 1):
        sim = cosine_similarity(bin_means[:, i].reshape(1, -1), bin_means[:, i+1].reshape(1, -1))[0,0]
        sims.append(sim)
    return np.array(sims)

def autocorr(x, lag=1):
    """Compute autocorrelation of 1D array at given lag."""
    n = len(x)
    x_mean = np.mean(x)
    denom = np.sum((x - x_mean)**2)
    if denom == 0:
        return 0
    num = np.sum((x[:n-lag] - x_mean) * (x[lag:] - x_mean))
    return num / denom

bin_size = 100

# Compute bin means and adjacent similarities
task_bin_means = mean_bin_embeddings(inputs_task, bin_size)
shuff_bin_means = mean_bin_embeddings(rest_shuff, bin_size)
rand_bin_means = mean_bin_embeddings(rest_random, bin_size)

task_adj_sims = adjacent_bin_similarity(task_bin_means)
shuff_adj_sims = adjacent_bin_similarity(shuff_bin_means)
rand_adj_sims = adjacent_bin_similarity(rand_bin_means)

# Variance
var_task = np.var(task_adj_sims)
var_shuff = np.var(shuff_adj_sims)
var_rand = np.var(rand_adj_sims)

# Autocorrelation at lag=1 (can test other lags)
autocorr_task = autocorr(task_adj_sims, lag=1)
autocorr_shuff = autocorr(shuff_adj_sims, lag=1)
autocorr_rand = autocorr(rand_adj_sims, lag=1)

# Plot variance and autocorrelation side by side
fig, axs = plt.subplots(1, 2, figsize=(10,4))

# Variance bar plot
axs[0].bar(['Task', 'Rest shuffled', 'Rest random'], [var_task, var_shuff, var_rand], color=['C0', 'C1', 'C2'])
axs[0].set_title("Variance of adjacent-bin similarity")
axs[0].set_ylabel("Variance")

# Autocorrelation bar plot
axs[1].bar(['Task', 'Rest shuffled', 'Rest random'], [autocorr_task, autocorr_shuff, autocorr_rand], color=['C0', 'C1', 'C2'])
axs[1].set_title("Autocorrelation (lag=1) of adjacent-bin similarity")
axs[1].set_ylabel("Autocorrelation")

plt.tight_layout()
plt.show()

print(f"Variance (Task, Shuffled, Random): {var_task:.4f}, {var_shuff:.4f}, {var_rand:.4f}")
print(f"Autocorrelation (Task, Shuffled, Random): {autocorr_task:.4f}, {autocorr_shuff:.4f}, {autocorr_rand:.4f}")
```
